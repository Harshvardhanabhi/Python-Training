```python
# Calculate Q1, Q3, and IQR
Q1 = df['money'].quantile(0.25)
Q3 = df['money'].quantile(0.75)
IQR = Q3 - Q1

# Identify outliers
df['outlier']= (df['money'] < (Q1 - 1.5 * IQR)) | (df['money'] > (Q3 + 1.5 * IQR))
```

```python
outliers_df = df[df['outlier']]
```
