```python
import pandas as pd
s = pd.Series(pd.to_datetime(['2025-07-23 14:30:00']))
print(s.dt.year)
print(s.dt.hour)
```

#   'D': Daily frequency
#    W': Weekly frequency
#    'M': Monthly frequency (end of month)
#    'H': Hourly frequency
#    'min' or 'T': Minute frequency
#    'S': Second frequency

```python
print(s.dt.minute)
```

```python
import pandas as pd

# Create a sample DataFrame with daily data
rng = pd.date_range('2025-01-01', periods=24, freq='D')  # Changed 'H' to 'D'
df=pd.DataFrame(rng)
```

```python
import numpy as np
dates = pd.date_range('2025-01-01', periods=100, freq='H')
data = np.random.randint(10,size=100)
df = pd.DataFrame({'value': data}, index=dates)
```

```python
df
```

```python
df
```

```python
mean=df.resample('D').mean()
```

```python
mean
```

```python
hourly_data = mean.resample('H').ffill()
```

```python
hourly_data
```

```python
import pandas as pd
import numpy as np

# Create a date range with hourly frequency for 3 days
rng = pd.date_range('2025-07-20 00:00:00', periods=72, freq='H')
```

```python
temperatures = np.random.randint(15, 30, size=len(rng))
```

```python
df=pd.DataFrame({'temperature' : temperatures, 'dates' : rng}).set_index('dates')
```

```python
df
```

```python
new= df.resample('D').mean()
```

```python
new
```

```python
up = df.resample('30m').ffill()
```
