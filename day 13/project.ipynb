```python
merge_df3['revenue'] = merge_df3['price']*merge_df3['quantity']
```

```python
merge_df3.head()
```

```python
store_wise = merge_df3.groupby(['store_name' , 'product'])['revenue'].agg(['sum' ])
```

```python
new_df= pd.DataFrame(store_wise)
```
