```python
import pandas as pd
df = pd.read_csv("sales_data_sample.csv", encoding='latin1',usecols=['ORDERDATE' ,'SALES'])
```

```python
df
```

```python
df['ORDER_DATE'] = pd.to_datetime(df['ORDERDATE'])
```

```python
new=df.drop('ORDERDATE', axis=1)
```

```python
roll_data = new['SALES'].rolling(window =7 ).mean()
```

```python
new['rolling_sales'] = roll_data
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
new_df =new.dropna()
new_df.drop('rolling' ,axis=1)
```
