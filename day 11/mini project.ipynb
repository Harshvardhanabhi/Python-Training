Mini Project: Generate daily, weekly, monthly sales summaries




```python
df = pd.read_csv("sales_data_sample.csv", encoding='latin1',usecols=['ORDERDATE' ,'MONTH_ID','YEAR_ID','COUNTRY','SALES'])
```

```python
df.head()
```

```python
a= df.pivot_table(index=['COUNTRY','YEAR_ID','MONTH_ID'] ,values='SALES' , aggfunc='sum')
```

```python
a.sort_values(by=['YEAR_ID','MONTH_ID'])
```
